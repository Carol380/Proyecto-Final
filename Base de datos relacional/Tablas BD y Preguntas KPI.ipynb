# Creación de CSVs en Google Drive:

```python
# Eliminar saltos de línea de df_contenido:

def replace_newlines(x):
  if isinstance(x, str):
    return x.replace('\n', ' ').replace('\r', ' ')
  return x

df_contenido['cont_texto'] = df_contenido['cont_texto'].apply(replace_newlines)

df_contenido.head()
```

```python
import csv

# Copiar los CSV a carpeta de Google Drive:
import os

# carpeta_destino = '/content/drive/My Drive/Proyecto Júpiter - Master DA & DS/Base de datos relacional/CSV para BD/'
carpeta_destino = "CSV para BD/"
nombre_archivo_noticias = 'Noticias.csv'
nombre_archivo_contenido = 'Contenido.csv'
nombre_archivo_tipo = 'Tipo.csv'
nombre_archivo_veracidad = 'Veracidad.csv'
nombre_archivo_pais_idioma = 'Pais_Idioma.csv'
nombre_archivo_fuente = 'Fuente.csv'
nombre_archivo_autor = 'Autor.csv'

# Crear la carpeta si no existe:
if not os.path.exists(carpeta_destino):
    os.makedirs(carpeta_destino)

csv_path_noticias = os.path.join(carpeta_destino, nombre_archivo_noticias)
df_noticias.to_csv(csv_path_noticias, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)

csv_path_contenido = os.path.join(carpeta_destino, nombre_archivo_contenido)
df_contenido.to_csv(csv_path_contenido, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)

csv_path_tipo = os.path.join(carpeta_destino, nombre_archivo_tipo)
df_tipo.to_csv(csv_path_tipo, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)

csv_path_pais_idioma = os.path.join(carpeta_destino, nombre_archivo_pais_idioma)
df_pais_idioma.to_csv(csv_path_pais_idioma, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)

csv_path_fuente = os.path.join(carpeta_destino, nombre_archivo_fuente)
df_fuente.to_csv(csv_path_fuente, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)

csv_path_autor = os.path.join(carpeta_destino, nombre_archivo_autor)
df_autor.to_csv(csv_path_autor, sep='|', index=False, quotechar='"', quoting=csv.QUOTE_ALL)
```

###• ¿Qué día del mes se han publicado más noticias, cuántas y de qué tipo son?

```python
# Día del mes con más noticias publicadas:
df_noticias['tiempo'] = pd.to_datetime(df_noticias['tiempo'], errors='coerce')

max_publicaciones = df_noticias.groupby([df_noticias['tiempo'].dt.month_name(), df_noticias['tiempo'].dt.day]).agg(cantidad_noticias = ('id_noticia','count')).sort_values(by='cantidad_noticias', ascending=False).head(1)
max_publicaciones
```

```python
# Por tipo:

# Filtrar día con más noticias publicadas:
mes_maximo, dia_maximo = max_publicaciones.index[0]
noticias_fecha_maxima = df_noticias[(df_noticias['tiempo'].dt.month_name() == mes_maximo) & (df_noticias['tiempo'].dt.day == dia_maximo)]

# Agrupar por tipo:
resultado_join = noticias_fecha_maxima.merge(df_tipo, on='id_tipo', how='left')
conteo_por_tipo = resultado_join.groupby('tipo').agg(cantidad_noticias=('id_noticia', 'count')).reset_index()
conteo_por_tipo = conteo_por_tipo.sort_values(by='cantidad_noticias', ascending=False)


print(f"Noticias por tipo en {mes_maximo} {dia_maximo}, ordenadas de mayor a menor:")
print(conteo_por_tipo)
```

###• ¿Qué días del mes se han compartido más noticias, cuántas y de qué tipo son?

```python
# Día del mes con más noticias compartidas:
df_noticias['compartir_tiempo'] = pd.to_datetime(df_noticias['compartir_tiempo'], errors='coerce')

max_compartido = df_noticias.groupby([df_noticias['compartir_tiempo'].dt.month_name(), df_noticias['compartir_tiempo'].dt.day]).agg(cantidad_noticias = ('id_noticia','count')).sort_values(by='cantidad_noticias', ascending=False).head(1)
max_compartido
```

```python
# Por tipo:

# Filtrar día con más noticias compartidas:
mes_maximo_comp, dia_maximo_comp = max_compartido.index[0]
noticias_fecha_maxima_comp = df_noticias[(df_noticias['compartir_tiempo'].dt.month_name() == mes_maximo_comp) & (df_noticias['compartir_tiempo'].dt.day == dia_maximo_comp)]

# Agrupar por tipo:
resultado_join_comp = noticias_fecha_maxima_comp.merge(df_tipo, on='id_tipo', how='left')
conteo_por_tipo_comp = resultado_join_comp.groupby('tipo').agg(cantidad_noticias=('id_noticia', 'count')).reset_index()
conteo_por_tipo_comp = conteo_por_tipo_comp.sort_values(by='cantidad_noticias', ascending=False)


print(f"Noticias por tipo en {mes_maximo_comp} {dia_maximo_comp}, ordenadas de mayor a menor:")
print(conteo_por_tipo_comp)
```

###• ¿Cuántas noticias publicadas han sido fake news?

```python
filtro_fake = df_noticias[df_noticias['real_fake_integer'] == 0] # 0 = FAKE
filtro_fake.groupby('real_fake_integer').agg(cantidad_noticias=('id_noticia','count'))
```

###• ¿Qué tipo de noticia tiene mayor porcentaje de ser fake new?

```python
# Noticias por tipo
conteo_total_por_tipo = df_noticias.groupby('id_tipo').size().reset_index(name='cantidad_total')

# Noticias fake por tipo
conteo_fake_por_tipo = filtro_fake.groupby('id_tipo').size().reset_index(name='cantidad_fake')

# Join con df_tipo:
join_tipo = conteo_total_por_tipo.merge(conteo_fake_por_tipo, on='id_tipo', how='left').merge(df_tipo, on='id_tipo', how='left')

# Calcular la proporción de noticias fake
join_tipo['proporcion_fake'] = (join_tipo['cantidad_fake'] / join_tipo['cantidad_total']) * 100

# Ordenar de mayor a menor por la proporción de noticias fake
resultado_ordenado = join_tipo.sort_values(by='proporcion_fake', ascending=False).head(1)

print("Tipo de noticia con mayor probabilidad de ser fake:")
print(resultado_ordenado[['tipo', 'proporcion_fake']])
```
